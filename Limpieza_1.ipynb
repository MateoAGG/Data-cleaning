```python
dt['crime_type'].unique()
# Poner todo en minuscula y eliminar espacios en blanco
dt['crime_type'] = dt['crime_type'].str.lower()
dt['crime_type'] = dt['crime_type'].str.strip()
# Crear un diccionario de correcciones para los valores mal escritos
corrections = {
    # Assault
    "asslt": "assault",
    "assault": "assault",
    # Assault and Battery
    "assault & battery": "assault and battery",
    "assault  &  battery": "assault and battery",
    "battery": "assault and battery",
    # Burglary / Breaking and Entering
    "burglary": "burglary",
    "burglry": "burglary",
    "breaking & entering": "burglary",
    "b&e": "burglary",
    # Homicide / Murder / Manslaughter
    "homocide": "homicide",
    "homicide": "homicide",
    "murder": "homicide",
    "manslaughter": "manslaughter",
    # Property Damage / Vandalism / Graffiti / Arson
    "property damage": "property damage",
    "property  damage": "property damage",
    "vandalism": "vandalism",
    "vandlism": "vandalism",
    "graffiti": "vandalism",
    "arson": "arson",
    "arsen": "arson",
    "fire setting": "arson",
    # Domestic Violence
    "domestc violence": "domestic violence",
    "domestic violence": "domestic violence",
    "domestic  violence": "domestic violence",
    "dom. violence": "domestic violence",
    "dv": "domestic violence",
    # Robbery
    "robbery": "robbery",
    "robbry": "robbery",
    "roberry": "robbery",
    "armed robbery": "armed robbery",
    "armed  robbery": "armed robbery",
    # Theft / Larceny
    "theft": "theft",
    "theft/larceny": "theft",
    "larceny": "theft",
    "stealing": "theft",
    # Drug Crimes
    "narcotics": "drug offense",
    "drug offense": "drug offense",
    "drug  offense": "drug offense",
    "drug offence": "drug offense",
    "drug  offence": "drug offense",
    "drugs": "drug offense",
    # Fraud / Scam / Deception
    "fraudulent activity": "fraud",
    "fraud": "fraud",
    "scam": "fraud",
    "deception": "fraud",
    "online fraud": "fraud",
    # Cyber Crime
    "cyber crime": "cyber crime",
    "cyber  crime": "cyber crime",
    "cybercrime": "cyber crime",
    "hacking": "cyber crime",
    # Sexual Assault
    "sexual assault": "sexual assault",
    "sexual  assault": "sexual assault",
    "sexual assualt": "sexual assault",
    "sexual  assualt": "sexual assault",
    "sex assault": "sexual assault",
    "sex  assault": "sexual assault",
    "sa": "sexual assault",
    # Kidnapping / Abduction
    "abduction": "kidnapping",
    "kidnaping": "kidnapping",
    "kidnapping": "kidnapping",
    # Trespassing
    "trespass": "trespassing",
    "trespassing": "trespassing",
    "tresspassing": "trespassing",
    # DUI / DWI
    "drunk driving": "dui",
    "duii": "dui",
    "d.u.i.": "dui",
    "dui": "dui",
    "dwi": "dui",
}
# Reemplazar los valores mal escritos con las correcciones
dt['crime_type'] = dt['crime_type'].replace(corrections)
# Verificar los valores únicos después de la limpieza
dt['crime_type'] = dt['crime_type'].str.capitalize()
display(dt['crime_type'].unique())
```

### District

```python
# Verificar Valores unicos en la columna 'district'
dt['district'].unique()
# Poner todo en minuscula y eliminar espacios en blanco
dt['district'] = dt['district'].str.lower()
dt['district'] = dt['district'].str.strip()
# Crear un diccionario de correcciones para los valores mal escritos
correciones = {
    'sou': 'south',
    'cen': 'central',
    'eas': 'east',
    'nor': 'north',
    'mid': 'midtown',
    'wes':'west'
}
# Reemplazar los valores mal escritos con las correcciones
dt['district'] = dt['district'].replace(correciones)
# Poner la primera letra en mayuscula
dt['district'] = dt['district'].str.capitalize()
dt['district'].unique()
```
