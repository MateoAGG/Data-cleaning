```python
import pandas as pd
```

```python
dt = pd.read_csv('Bases\crime_incidents_messy.csv')
```

## Limpieza de variables

### Crime Type

```python
# Verificar tipo de datos
dt['crime_type'].dtype
# Verificar los valores nulos en la columna 'crime_type'
display(dt['crime_type'].isnull().sum())
# Verificar los valores únicos antes de la limpieza
dt['crime_type'].unique()
# Poner todo en minuscula y eliminar espacios en blanco
dt['crime_type'] = dt['crime_type'].str.lower()
dt['crime_type'] = dt['crime_type'].str.strip()
# Crear un diccionario de correcciones para los valores mal escritos
corrections = {
    # Assault
    "asslt": "assault",
    "assault": "assault",
    # Assault and Battery
    "assault & battery": "assault and battery",
    "assault  &  battery": "assault and battery",
    "battery": "assault and battery",
    # Burglary / Breaking and Entering
    "burglary": "burglary",
    "burglry": "burglary",
    "breaking & entering": "burglary",
    "b&e": "burglary",
    # Homicide / Murder / Manslaughter
    "homocide": "homicide",
    "homicide": "homicide",
    "murder": "homicide",
    "manslaughter": "manslaughter",
    # Property Damage / Vandalism / Graffiti / Arson
    "property damage": "property damage",
    "property  damage": "property damage",
    "vandalism": "vandalism",
    "vandlism": "vandalism",
    "graffiti": "vandalism",
    "arson": "arson",
    "arsen": "arson",
    "fire setting": "arson",
    # Domestic Violence
    "domestc violence": "domestic violence",
    "domestic violence": "domestic violence",
    "domestic  violence": "domestic violence",
    "dom. violence": "domestic violence",
    "dv": "domestic violence",
    # Robbery
    "robbery": "robbery",
    "robbry": "robbery",
    "roberry": "robbery",
    "armed robbery": "armed robbery",
    "armed  robbery": "armed robbery",
    # Theft / Larceny
    "theft": "theft",
    "theft/larceny": "theft",
    "larceny": "theft",
    "stealing": "theft",
    # Drug Crimes
    "narcotics": "drug offense",
    "drug offense": "drug offense",
    "drug  offense": "drug offense",
    "drug offence": "drug offense",
    "drug  offence": "drug offense",
    "drugs": "drug offense",
    # Fraud / Scam / Deception
    "fraudulent activity": "fraud",
    "fraud": "fraud",
    "scam": "fraud",
    "deception": "fraud",
    "online fraud": "fraud",
    # Cyber Crime
    "cyber crime": "cyber crime",
    "cyber  crime": "cyber crime",
    "cybercrime": "cyber crime",
    "hacking": "cyber crime",
    # Sexual Assault
    "sexual assault": "sexual assault",
    "sexual  assault": "sexual assault",
    "sexual assualt": "sexual assault",
    "sexual  assualt": "sexual assault",
    "sex assault": "sexual assault",
    "sex  assault": "sexual assault",
    "sa": "sexual assault",
    # Kidnapping / Abduction
    "abduction": "kidnapping",
    "kidnaping": "kidnapping",
    "kidnapping": "kidnapping",
    # Trespassing
    "trespass": "trespassing",
    "trespassing": "trespassing",
    "tresspassing": "trespassing",
    # DUI / DWI
    "drunk driving": "dui",
    "duii": "dui",
    "d.u.i.": "dui",
    "dui": "dui",
    "dwi": "dui",
}
# Reemplazar los valores mal escritos con las correcciones
dt['crime_type'] = dt['crime_type'].replace(corrections)
# Verificar los valores únicos después de la limpieza
dt['crime_type'] = dt['crime_type'].str.capitalize()
display(dt['crime_type'].unique())
```

### District

```python
# Verificar tipo de datos
dt['district'].dtype
# Verificar los valores nulos en la columna 'district'
display(dt['district'].isnull().sum())
# Verificar Valores unicos en la columna 'district'
dt['district'].unique()
# Poner todo en minuscula y eliminar espacios en blanco
dt['district'] = dt['district'].str.lower()
dt['district'] = dt['district'].str.strip()
# Crear un diccionario de correcciones para los valores mal escritos
correciones = {
    'sou': 'south',
    'cen': 'central',
    'eas': 'east',
    'nor': 'north',
    'mid': 'midtown',
    'wes':'west'
}
# Reemplazar los valores mal escritos con las correcciones
dt['district'] = dt['district'].replace(correciones)
# Poner la primera letra en mayuscula
dt['district'] = dt['district'].str.capitalize()
dt['district'].unique()
```

### City

```python
# Verificar tipo de datos
dt['city'].dtype
# Verificar los valores nulos en la columna 'city'
display(dt['city'].isnull().sum())
dt['city'].unique()
dt['city'] = dt['city'].str.strip()
dt['city'].unique()
```

### State

```python
# Verificar tipo de datos
dt['state'].dtype
# Verificar los valores nulos en la columna 'state'
display(dt['state'].isnull().sum())
# Verificar los valores únicos en la columna 'state'
dt['state'].unique()
```

### Adress

```python
# Verificar tipo de datos
dt['address'].dtype
# Verificar los valores nulos en la columna 'address'
display(dt['address'].isnull().sum())
# Verificar los valores únicos en la columna 'address'
dt['address'].unique()
# Separar la dirección en calle y número
dt[['number','street']] = dt['address'].str.split(' ',expand=True, n=1)
# Verificar los valores nulos, uniques y el resultado de la separación
display(dt['number'].isnull().sum())
display(dt['street'].isnull().sum())
display(dt['number'].unique())
display(dt['street'].unique())
```

### Latitude - Longitude

```python
# Verificar valores nulos en las columnas 'latitude' y 'longitude'
dt['latitude'].dtype
dt['longitude'].dtype
# Verificar los valores nulos en las columnas 'latitude' y 'longitude'
nulos = dt['latitude'].isnull().sum() + dt['longitude'].isnull().sum()
total_filas = len(dt)
porcentaje_nulos = (nulos / total_filas) * 100
print(f"Total: {porcentaje_nulos:.2f}% de los datos tienen valores nulos en 'latitude' o 'longitude'.")
# Se debe eliminar estos valores nulos para evitar problemas en el análisis geoespacial
dt = dt.dropna(subset=['latitude', 'longitude'])
```

### Incident Datetime

```python
# Verificar tipo de datos
dt['incident_datetime'].dtype
# verificar los valores nulos en la columna 'incident_datetime'
display(dt['incident_datetime'].isnull().sum())
# Dividir hora y Fecha
dt[['hour','Date']] = dt['incident_datetime'].str.split(' ',expand=True, n=1)
```

### Officer ID, FirstName  and LastName 

```python
# Verificar tipo de datos
display(dt['officer_id'].dtype)
display(dt['officer_first_name'].dtype)
display(dt['officer_last_name'].dtype)
# Verificar los valores nulos en las columnas 'officer_id', 'officer_first_name' y 'officer_last_name'
nulos_oficiales = dt['officer_id'].isnull().sum() + dt['officer_first_name'].isnull().sum() + dt['officer_last_name'].isnull().sum()
print(f"Total: {nulos_oficiales} valores nulos en las columnas de oficiales")
```

### Badget Number

```python
# Verificar tipo de datos
display(dt['badge_number'].dtype)
#Deberia ser int, pero hay valores no numéricos que deben ser corregidos o eliminados
display(dt['badge_number'].isnull().sum())
#Convertir a tipo numérico
dt['badge_number'] = dt['badge_number'].astype('Int64')
```
