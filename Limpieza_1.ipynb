# Limpieza de datos

```python
import pandas as pd
```

```python
dt = pd.read_csv('Bases\crime_incidents_messy.csv')
```

## Limpieza de variables

### Crime Type

```python
dt['crime_type'].unique()
# Poner todo en minuscula y eliminar espacios en blanco
dt['crime_type'] = dt['crime_type'].str.lower()
dt['crime_type'] = dt['crime_type'].str.strip()
dt['crime_type'].unique()
```

### District

```python
# Verificar Valores unicos en la columna 'district'
dt['district'].unique()
# Poner todo en minuscula y eliminar espacios en blanco
dt['district'] = dt['district'].str.lower()
dt['district'] = dt['district'].str.strip()
# Crear un diccionario de correcciones para los valores mal escritos
correciones = {
    'sou': 'south',
    'cen': 'central',
    'eas': 'east',
    'nor': 'north',
    'mid': 'midtown',
    'wes':'west'
}
# Reemplazar los valores mal escritos con las correcciones
dt['district'] = dt['district'].replace(correciones)
# Poner la primera letra en mayuscula
dt['district'] = dt['district'].str.capitalize()
dt['district'].unique()
```
